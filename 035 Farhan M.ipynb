```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from selenium import webdriver
import chromedriver_autoinstaller
import csv
import time
```

```python
#INSTALL BROWSER SUCH CHROME OR EDGE
chromedriver_autoinstaller.install()

#USE THE INSTALLED BROWSER FOR SELENIUM
browser = webdriver.Chrome()
```

```python
data = []

for index in range (1,3):
    browser.get(f'https://www.tokopedia.com/search?navsource=&page={index}&q=sempak&srp_component_id=02.01.00.00&srp_page_id=&srp_page_title=&st=product')
    time.sleep(10)
    soup = BeautifulSoup(browser.page_source,"lxml")
    rows = soup.find_all('div', class_='css-llwpbs')
```

```python
    for row in rows:
        nama = row.find(class_='prd_link-product-name css-3um8ox')
        if nama:
            nama = nama.text
            
        sold = row.find(class_='prd_label-integrity css-1duhs3e')
        if sold:
            sold = sold.text
            
        rating = row.find(class_='prd_rating-average-text css-t70v7i')
        if rating:
            rating = rating.text
            
        Discount = row.find(class_='prd_badge-product-discount css-1qtulwh')
        if Discount:
            Discount = Discount.text
        
        cashback = row.find(class_='prd_label-product-price css-tolj34')
        if cashback:
            cashback = cashback.text
        
        price = row.find(class_='prd_link-product-price css-1ksb19c')
        if price:
            price = price.text
        
        seller_loct = row.find(class_='prd_link-shop-loc css-1kdc32b flip')
        if seller_loct:
            seller_loct = seller_loct.text
            
        
            
        #Nama_Toko = row.find(class_='prd_link-shop-name css-1kdc32b flip').text
        #Lokasi_Penjual = row.find(class_='prd_link-shop-loc css-1kdc32b flip').text
        
        
        df = {
            'Nama': nama,
            'Terjual': sold,
            'Rating': rating,
            'Diskon': Discount,
            'Cashback': cashback,
            'Harga': price,
            'Lokasi Toko': seller_loct
            }
        data.append(df)
```

```python
df = pd.DataFrame(data)
```
